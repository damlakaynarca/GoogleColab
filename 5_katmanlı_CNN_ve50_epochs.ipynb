```python
import tensorflow as tf

# MNIST veri setini yükle
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0  # Verileri 0-1 aralığında normalize et

# Modeli oluştur
model = tf.keras.models.Sequential([
    tf.keras.layers.Reshape((28, 28, 1), input_shape=(28, 28)),  # Giriş katmanı: 28x28 boyutunda, kanal sayısı 1 olan bir görüntü
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),       # Convolutional katman: 32 filtre, 3x3 boyutunda, ReLU aktivasyonu
    tf.keras.layers.MaxPooling2D((2, 2)),                        # MaxPooling katmanı: 2x2 boyutunda
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),       # Convolutional katman: 64 filtre, 3x3 boyutunda, ReLU aktivasyonu
    tf.keras.layers.MaxPooling2D((2, 2)),                        # MaxPooling katmanı: 2x2 boyutunda
    tf.keras.layers.Flatten(),                                   # Düzleştirme katmanı
    tf.keras.layers.Dense(128, activation='relu'),               # Tam bağlı gizli katman: 128 nöronlu, ReLU aktivasyonu
    tf.keras.layers.Dense(10, activation='softmax')              # Çıkış katmanı: 10 sınıf için softmax aktivasyonu
])

# Modeli derle
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Modeli eğit
model.fit(x_train, y_train, epochs=50)  # 50 epoch

# Modelin performansını değerlendir
test_loss, test_acc = model.evaluate(x_test, y_test)
print('\nTest accuracy:', test_acc)

```
